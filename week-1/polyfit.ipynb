```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def noise(y_base, low=0, high=0.2):
  sign = np.random.uniform(-1, 1, len(y_base))
  scale = np.random.uniform(low, high, len(y_base))
  return sign * scale * y_base
```

```python
x = np.arange(0, 20)

# params: y = mx + b
m_true = 5
b_true = 2

# actual output
y_base = m_true * x + b_true
y = y_base + noise(y_base)

# predicted output
m, b = np.polyfit(x, y, deg=1)
y_pred = m * x + b

_ = plt.scatter(x, y, color='red')  # data
_ = plt.plot(x, y_pred)  # regression curve
```

```python
x = np.arange(-10, 10)

# params: y = a1x**2 + a2x + c
a1_true = 1
a2_true = 0
c_true = 0

# actual output
y_base = (a1_true*x**2) + (a2_true*x) + c_true
y = y_base + noise(y_base)

# predicted output
a1, a2, c = np.polyfit(x, y, deg=2)
y_pred = (a1*x**2) + (a2*x) + c

_ = plt.scatter(x, y, color='red')  # data
_ = plt.plot(x, y_pred)  # regression curve
```

```python
x = np.arange(-10, 10)

# params: y = a1x**3 + a2x**2 + a3x + c
a1_true = 1
a2_true = 0
a3_true = 0
c_true = 0

# actual output
y_base = (a1_true*x**3) + (a2_true*x**2) + (a3_true*x) + c_true
y = y_base + noise(y_base)

# predicted output
a1, a2, a3, c = np.polyfit(x, y, deg=3)
y_pred = (a1*x**3) + (a2*x**2) + (a3*x) + c

_ = plt.scatter(x, y, color='red')  # data
_ = plt.plot(x, y_pred)  # regression curve
```
